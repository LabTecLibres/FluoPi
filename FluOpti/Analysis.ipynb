```python
def f_gompertz(t, a, b, c):
    """
    Compute the gompertz function value using the given input values
    
    Parameters
    ----------
        t: vector
            independent variable ( "x axis", suposed to be time) 
        
        a: double
            maximum growth value parameter
        
        b: double
            maximum growth rate parameter
            
        c: double
            delay parameter
        
    Returns
    -------
    function evaluation
    
    """
    return(a*np.exp(-np.exp(1 + b*np.exp(1)*(c-t)/a)))  
```

```python
R_Mean_int = database[experiments[0]][series[2]]['Fluorescence Intensity']
r_colonies = database[experiments[0]][series[1]]['Selected colonies']
T_r = database[experiments[0]][series[0]]
```

```python
RF_path = 'Plots/'+experiments[0]+'//Fluo'
```

```python
#set the color vector to have concordance between dots and fitted curves
colors = plt.cm.jet(np.linspace(0, 1, len(r_colonies))) #Paired o jet #https://matplotlib.org/users/colormaps.html
```

```python
#let's sum the three channels
MeanInt3C={}
for i in r_colonies:
    Vaux=np.zeros((R_Mean_int['R'][i].shape))
    for c in channels:
        Vaux+=R_Mean_int[c][i]
    
    MeanInt3C[i]= Vaux
```

```python
from scipy.interpolate import UnivariateSpline
```

```python
IA_smooth={}

k_spline_IA = 4  # k is the degree of the smoothing spline  (<=5)

for i, c in zip(r_colonies, colors):
    IA_smooth[i] = UnivariateSpline(T_r,MeanInt3C[i], k=k_spline_IA )   
    IA_smooth[i].set_smoothing_factor(10)
    plt.plot(T_r,IA_smooth[i](T_r[:]),'-',color=c, label='colony '+str(i))
    plt.plot(T_r,MeanInt3C[i],'x',color=c)
    plt.xlabel('Time (h)')
    plt.ylabel('I/A $[RGB \cdot pixels^{-2}]$')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    #plt.legend(loc='upper left')
    plt.ylim([-3,90])

plt.savefig(RF_path+"//IA_smooth_axis.pdf", transparent=True) # Uncomment to save image
#plt.savefig(F_path+"//IA_smooth_labels.pdf", transparent=True) # Uncomment to save image
```

```python
R_Kmax = {}
limits = [10,50]

for i in r_colonies:
    Kf_values = R_K[i][limits[0]:limits[1]]
    
    max_val = Kf_values.max()
    time_point = np.where( Kf_values == max_val)
    
    R_Kmax[i] = [max_val, time_point[0][0]]
    
```

```python
G_Kmax = {}
limits = [10,50]

for i in g_colonies:
    Kf_values = G_K[i][limits[0]:limits[1]]
    
    max_val = Kf_values.max()
    time_point = np.where( Kf_values == max_val)
    
    G_Kmax[i] = [max_val, time_point[0][0]]
```

```python
plt.figure()
r_values = []
g_values = []
for i in r_colonies:
    plt.plot(1,R_Kmax[i][0], 'ro')
    r_values.append(R_Kmax[i][0])

for i in g_colonies:
    plt.plot(2,G_Kmax[i][0], 'go')
    g_values.append(G_Kmax[i][0])

plt.ylabel('K_max')
plt.xlim([0.5,2.5])
plt.xticks([1,2],['Red Light','Green Light'])
```
